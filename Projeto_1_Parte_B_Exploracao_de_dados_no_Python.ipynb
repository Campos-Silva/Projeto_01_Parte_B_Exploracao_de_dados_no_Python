# 1 - Análises exploratórias para as variáveis numéricas

###  -  Detectando e lidando com outliers

- Outliers podem induzir a erros nos modelos. Assim, para esse projeto irei primeiro identificar e lidar com eles para as variáveis numéricas.

#### Variável numérica nº 1: "preco_de_venda_USD"

```
# Visualizando os dados através de boxplot

plt.figure(figsize=(10,5))

sns.boxplot(x=carros.preco_de_venda_USD)

plt.xlabel('Preço de veículos usados (USD)', fontsize = 15)
```

#### **Conclusão prévia:**

- Por meio desse gráfico é possível ver que há valores que se distanciam muito dos outros para essa variável. Esses valores podem ser considerados possivelmente outliers.

- Há funções matemáticas que nos permitem identificar precisamente tais outliers, tal como o Score Z e o "Interquartile range (IQR)".

- Como a proposta da Parte B é demonstrar como fazemos uma análise de exploração de dados eu vou remover esses outliers a fim de demonstrar como realizamos tal tarefa. Irei fazer isso utilizando o Interquartile range (IQR).



```
for x in ['preco_de_venda_USD']:
    q75,q25 = np.percentile(carros.loc[:,x],[75,25])
    intr_qr = q75-q25
 
    max = q75+(1.5*intr_qr)
    min = q25-(1.5*intr_qr)
 
    carros.loc[carros[x] < min,x] = np.nan
    carros.loc[carros[x] > max,x] = np.nan
```

```
# Identificando quantos valores nulos há no dataset por categoria

carros.isnull().sum()
```

```
# Removendo as linhas que possuem valores nulos

carros = carros.dropna(axis = 0)
```

```
# Vou identificar se existem valores nulos por categorias

carros.isnull().sum()
```

```
# Revendo se existe valores discrepantes

plt.figure(figsize=(10,5))

sns.boxplot(x=carros.preco_de_venda_USD)

plt.xlabel('Preço de veículos usados (USD)', fontsize = 15)
```

```
carros.shape
```

#### **Conclusão prévia:**

- Foi possível notar que houve uma acentuada diminuição de valores discrepantes para essa variável. Vou prosseguir para as outras variáveis a fim de realizar a detecção de outliers.

### 1 - Variável: "tipo_combustivel"

Criando um dataframe para a variável e identificando a quantidade de valores por meio da função "value_counts()"

```
#Vou inserir um índice para criar duas colunas nesse dataframe.

carros.tipo_combustivel.value_counts().to_frame().reset_index()
```

```
# Vou renomear esse produto

tipo_combustivel = carros.tipo_combustivel.value_counts().to_frame().reset_index()

tipo_combustivel.head()
```

```
#Checando os valores por categoria

tipo_combustivel
```

```
# Colocando por ordem crescente de valores 

tipo_combustivel_sorted_cresc = tipo_combustivel.sort_values('Quantidade_de_carros', ascending=True)

tipo_combustivel_sorted_cresc
```

Gráfico

```
# Agora sim vou criar um gráfico para essa variável

# Vou utilizar um gráfico de barras da biblioteca Matplotlib

#_____________________________________________________________________

plt.figure(figsize=(12,6))

# Definindo os valores

plt.barh('Tipo_combustivel', 'Quantidade_de_carros', data=tipo_combustivel_sorted_cresc)

#Colocando os valores dentro de cada barra

for index, value in enumerate(tipo_combustivel_sorted_cresc['Quantidade_de_carros']):
    plt.text(value, index, str(value))

# Remover o eixo y

ax = plt.gca()
ax.axes.xaxis.set_visible(False)

# titulo e legenda dos eixos

plt.title('Há mais veículos a gasolina do que diesel', fontsize = 15)
plt.xlabel('Número de veículos')

plt.show()
```

## **Conclusão prévia:**

- Há mais carros a gasolina do que diesel nesse conjunto de dados.

### 1 - O preço do veículo está relacionado com o tipo de combustível?

Checando visualmente possiveis diferenças de valores de preço dos veículos usados por tipo de combustível.

```
# Boxplot de preço de venda por tipo de combustível

sns.boxplot(x = "tipo_combustivel", y = "preco_de_venda_USD", data=carros)
```

Parece haver diferença de valores de preço por tipo de combustível. No entanto, para saber se isso de fato é real preciso fazer um teste estatistico.

Mas, antes disso preciso saber qual é o tipo de distribuição de preços para cada tipo de combustivel. Isso será importante para determinar se irei realizar um teste do tipo paramétrico (se as distribuições forem do tipo normal), ou não paramétrico (caso não sejam normais).

Vou criar um dataframe de preço por tipo de combustível e logo após vou separar cada valor de preço por tipo de combustível.

```
# Criando o dataframe

carros_combustivel = carros[["tipo_combustivel", "preco_de_venda_USD"]]

carros_combustivel
```

Agora vou separar o valor de preço para cada tipo de combustível. Cada um ficará em um dataframe específico.

Vou fazer um filtro para selecionar o preço apenas de Diesel

```
# Dataframe de Diesel

p_Diesel = carros_combustivel[carros_combustivel['tipo_combustivel'] == 'Diesel']

p_Diesel
```

Vou fazer um filtro para selecionar o preço apenas de Gasolina

```
# Data frame de Gasolina

p_Gasolina = carros_combustivel[carros_combustivel['tipo_combustivel'] == 'Gasolina']

p_Gasolina
```

### Checando visualmente as distribuições do preco de veículos por tipos de combustíveis

```
# Montando o histograma

import matplotlib.pyplot as plt

fig, (ax1, ax2) = plt.subplots(1, 2)

fig.suptitle('Histograma de frequência do preço por tipo de combustível')

ax1.hist(p_Diesel['preco_de_venda_USD'], bins=10, histtype='bar', ec='k') 

ax2.hist(p_Gasolina['preco_de_venda_USD'], bins=10, histtype='bar', ec='k') 
ax1.set_xlabel("Preço - Diesel")
ax2.set_xlabel("Preço - Gasolina")
plt.show()
```

A partir desses gráficos é possível diagnosticar que, provavelmente, os valores de preço por tipo de combustíveis **não** pertencem ao tipo de distribuição normal, isso para ambos os casos.

Para checar se isso é verdadeiro preciso realizar o teste de Shapiro-Wilk. Esse teste demonstra se a distribuição pertence ao tipo normal (quando tem valor de p maior que 0,05) ou não (quando p é menor que 0,05).

```
# Checando o tipo de distribuição para "Diesel"

w, pvalue = stats.shapiro(p_Diesel['preco_de_venda_USD'])
w, pvalue
```

O valor de p é menor que 0,05 e isso indica que o preço para veículos a diesel de fato **não** tem distribuição do tipo normal.

```
# Checando o tipo de distribuição para "Gasolina"

w, pvalue = stats.shapiro(p_Gasolina['preco_de_venda_USD'])
w, pvalue
```

O valor de p é menor que 0,05 e isso indica que o preço para veículos a gasolina de fato **não** tem distribuição do tipo normal.

Como as distribuições são do tipo "**não-normal**", tenho de realizar um teste estatístico não paramétrico para checar diferenças significativas entre as medianas. Para isso vou utilizar o teste de "Mann-Whitney". Caso o valor de p seja **menor** que 0,05 isso indicará que existe diferença entre os grupos. 

```
# Importando a biblioteca

from scipy.stats import mannwhitneyu
```

```
# Carrying out the Wilcoxon–Mann–Whitney test

results = mannwhitneyu(p_Diesel['preco_de_venda_USD'], p_Gasolina['preco_de_venda_USD'])
results
```

## **Conclusão**

o valor de p é menor que 0,05. Assim, **existe diferença de valores entre as medianas para preço de veículos com relação ao tipo de combustível**. Dessa forma, eu irei manter essa variável no dataset para construir o modelo preditivo na Parte C do projeto.

Vou fazer um filtro para selecionar o preço apenas de "**Automático**".

```
# Data frame de Automático

p_Automatico = carros_transmissao[carros_transmissao['transmissao'] == 'Automatico']

p_Automatico
```

### Checando visualmente as distribuições do preco de veículos por tipos de Transmissão

```
# Montando o histograma

import matplotlib.pyplot as plt

fig, (ax1, ax2) = plt.subplots(1, 2)

fig.suptitle('Histograma de frequência do preço por tipo de transmissão')

ax1.hist(p_Manual['preco_de_venda_USD'], bins=10, histtype='bar', ec='k') 

ax2.hist(p_Automatico['preco_de_venda_USD'], bins=10, histtype='bar', ec='k')


ax1.set_xlabel("Preço - Manual")
ax2.set_xlabel("Preço - Automatico")
plt.show()
```

A partir desses gráficos é possível diagnosticar que, provavelmente, os valores de preço por tipo de transmissão **não** pertencem ao tipo de distribuição normal, isso para ambos os casos.

Para checar se isso é verdadeiro preciso realizar o teste de Shapiro-Wilk. Esse teste demonstra se a distribuição pertence ao tipo normal (quando tem valor de p maior que 0,05) ou não (quando p é menor que 0,05).

```
# Checando o tipo de distribuição para "Manual"

w, pvalue = stats.shapiro(p_Manual['preco_de_venda_USD'])
w, pvalue
```

O valor de p é menor que 0,05 e isso indica que o preço para veículos com **transmissão manual** de fato **não** tem distribuição do tipo normal.

```
# Checando o tipo de distribuição para "Automatico"

w, pvalue = stats.shapiro(p_Automatico['preco_de_venda_USD'])
w, pvalue
```

O valor de p é menor que 0,05 e isso indica que o preço para veículos com **transmissão automática** de fato **não** tem distribuição do tipo normal.

Como as distribuições são do tipo "**não-normal**", tenho de realizar um teste estatístico não paramétrico para checar diferenças significativas entre as medianas. Para isso vou utilizar o teste de "Mann-Whitney". Caso o valor de p seja **menor** que 0,05 isso indicará que existe diferença entre os grupos. 

```
# Importando a biblioteca

from scipy.stats import mannwhitneyu
```

```
# Carrying out the Wilcoxon–Mann–Whitney test

results = mannwhitneyu(p_Manual['preco_de_venda_USD'], p_Automatico['preco_de_venda_USD'])
results
```

## **Conclusão**

o valor de p é menor que 0,05. Assim, **existe diferença de valores entre as medianas para preço de veículos com relação ao tipo de transmissão**. Dessa forma, eu irei manter essa variável no dataset para construir o modelo preditivo na Parte C do projeto.

A partir desses gráficos é possível diagnosticar que, provavelmente, os valores de preço por quantidade prévia de donos **não** pertencem ao tipo de distribuição normal (com exceção de "Distribuidora").

Para checar se isso é verdadeiro preciso realizar o teste de Shapiro-Wilk. Esse teste demonstra se a distribuição pertence ao tipo normal (quando tem valor de p maior que 0,05) ou não (quando p é menor que 0,05).

```
# Checando o tipo de distribuição para "1º dono"

w, pvalue = stats.shapiro(p_1_dono['preco_de_venda_USD'])
w, pvalue
```

O valor de p é menor que 0,05 e isso indica que o preço para veículos que possuem apenas 1 dono de fato **não** tem distribuição do tipo normal.

```
# Checando o tipo de distribuição para "2º dono"

w, pvalue = stats.shapiro(p_2_dono['preco_de_venda_USD'])
w, pvalue
```

O valor de p é menor que 0,05 e isso indica que o preço para veículos que possuem 2 donos de fato **não** tem distribuição do tipo normal.

```
# Checando o tipo de distribuição para "3º dono"

w, pvalue = stats.shapiro(p_3_dono['preco_de_venda_USD'])
w, pvalue
```

O valor de p é menor que 0,05 e isso indica que o preço para veículos que possuem 3 donos de fato **não** tem distribuição do tipo normal.

```
# Checando o tipo de distribuição para "4º dono e acima"

w, pvalue = stats.shapiro(p_4_dono['preco_de_venda_USD'])
w, pvalue
```

O valor de p é menor que 0,05 e isso indica que o preço para veículos que possuem 4 ou mais donos de fato **não** tem distribuição do tipo normal.

Como as distribuições são do tipo "**não-normal**", tenho de realizar um teste não paramétrico para checar diferenças significativas entre as medianas. Para isso vou utilizar o teste de "Kruskal-Wallis", uma vez que são tres grupos a serem comparados. Caso o valor de p seja **menor** que 0,05 isso indicará que existe diferença entre os grupos.

```
# Importando a biblioteca

from scipy import stats
```

```
# Fazendo o teste de Kruskal-Wallis

results = stats.kruskal(p_1_dono['preco_de_venda_USD'],
                        p_2_dono['preco_de_venda_USD'],
                        p_3_dono['preco_de_venda_USD'],
                        p_4_dono['preco_de_venda_USD'])
results
```
